```python
import netCDF4 as nc

# Carga el archivo .nc
archivo = '/humidity.nc'
ds = nc.Dataset(archivo)

# Explora las dimensiones y variables
print(ds)

# Para ver una lista de las variables del archivo
print(ds.variables.keys())

```

```python
import pandas as pd
import numpy as np
```

```python
humidity = ds.variables['r'][:]  # Tu array de humedad
date = ds.variables['date'][:]  # Tu array de fechas
```

```python
fechas_legibles = pd.to_datetime(date.astype(str), format='%Y%m%d')

# Aplanar la variable de humedad para que coincida con las fechas
humidity_flat = humidity.reshape(humidity.shape[0], -1)

# Crear un DataFrame con las fechas y los valores de humedad
df = pd.DataFrame(humidity_flat, index=fechas_legibles)

# Renombrar las columnas del DataFrame si es necesario
df.columns = [f'Humedad_{i+1}' for i in range(humidity_flat.shape[1])]

# Mostrar el DataFrame resultante
print(df)
```

```python
# Elige el índice de latitud y longitud que desees
lat_index = 0  # Cambia esto al índice de latitud que desees (0-18)
lon_index = 0  # Cambia esto al índice de longitud que desees (0-16)

# Seleccionar la humedad en el punto específico (con solo un nivel de presión)
humedad_punto = ds.variables['r'][:, 0, lat_index, lon_index]  # [date, pressure_level, latitude, longitude]

# Convertir a un formato unidimensional
humedad_punto = humedad_punto.flatten()

# Crear el DataFrame
df = pd.DataFrame({
    'Fecha': fechas_legibles,
    'Humedad': humedad_punto
})

# Asegúrate de que la longitud de 'Humedad' y 'Fecha' coincida
print("Tamaño de Humedad:", humedad_punto.shape)
print("Tamaño de Fechas:", len(fechas_legibles))

# Establecer la fecha como índice del DataFrame
df.set_index('Fecha', inplace=True)

# Mostrar el DataFrame
print(df)
```

```python
# Seleccionar la humedad para todas las latitudes y longitudes
humedad_area = ds.variables['r'][:, 0, :, :]  # Esto selecciona todas las fechas, 1 nivel de presión, todas las latitudes y longitudes

# Verifica la forma del array de humedad
print("Forma de humedad_area:", humedad_area.shape)  # Debe ser (372, 19, 17)

# Promedia sobre las dimensiones de latitud y longitud (2 y 3)
humedad_promedio = humedad_area.mean(axis=(1, 2))  # Promedio sobre latitud y longitud
humedad_promedio = humedad_promedio.flatten()  # Esto convierte la humedad a un array unidimensional

# Crear el DataFrame para el promedio
df_promedio = pd.DataFrame({
    'Fecha': fechas_legibles,
    'Humedad_Promedio': humedad_promedio
})

# Establecer la fecha como índice
df_promedio.set_index('Fecha', inplace=True)

# Mostrar el DataFrame
print(df_promedio)
```

```python
# Guardar el DataFrame en un archivo de texto (.txt)
df_promedio.to_csv('humedad_promedio.txt', sep='\t', index=True)  # sep='\t' para tabulaciones
```
